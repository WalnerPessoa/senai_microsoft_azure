One Dimension?
==============

We can break down the reliability test into two stages. In the first
stage, we can ask whether we are measuring "one thing" or several things
with this set of questions. *After all, you cannot have a reliable
measure of something unless you know you are measuring one thing to
begin with.* Although many people skip this step, I illustrate it here.

A factor analysis (similar to a "principal component analysis," which
you will learn later) can tell us how many underlying dimensions appear
to be operating beneath this set of variables. For example, if I asked
someone feeling angry if they feel "upset," "mad," and "angry" ... I
would likely get similar answers because these questions are all
assessing the **one** underlying dimension of anger. There does not
appear to be a separate dimension beneath those items (we call questions
"items" in survey writing). Similarly, we need to know if this set of
questions we have asked represents one underlying dimension or several.
We call these underlying dimensions "factors." Thus, we need to know how
many factors we have.
___

Uma dimensão?
===
Podemos dividir o teste de confiabilidade em dois estágios. No primeiro estágio, podemos perguntar se estamos medindo "uma coisa" ou várias coisas com esse conjunto de perguntas. Afinal, você não pode ter uma medida confiável de algo a menos que saiba que está medindo algo para começar. Embora muitas pessoas pulem esta etapa, eu a ilustro aqui.

Uma análise fatorial (semelhante a uma "análise de componente principal", que você aprenderá mais tarde) pode nos dizer quantas dimensões subjacentes parecem estar operando sob esse conjunto de variáveis. Por exemplo, se eu perguntasse a alguém que está com raiva se ela se sente "chateada", "descontrolada" e "brava" ... Provavelmente, obteria respostas semelhantes, porque todas essas perguntas avaliam a dimensão subjacente da raiva. Não parece haver uma dimensão separada abaixo desses itens (chamamos as perguntas de "itens" na escrita da pesquisa). Da mesma forma, precisamos saber se este conjunto de perguntas que fizemos representa uma dimensão subjacente ou várias. Chamamos essas dimensões subjacentes de "fatores". Portanto, precisamos saber quantos fatores temos.

___

I won't explain the details of factor analysis here (beyond the scope of
this course), but it essentially analyzes the patterns of association
between the items to determine if they "cluster together" and if so,
how. We start with `fa.parallel()` from the `psych` package. We tell it
we want to use the `minres` factor method (`fm = 'minres'`), which makes
few assumptions about the normality of the data, and we tell it we want
the analysis to be a "factor analysis" (`fa = 'fa'`).

I won't go into details about the analysis, but we compute a number of
"eigenvalues" for our data. Historically, the number of eigenvalues
above 1.0 was seen as the number of factors in the data. For example, if
we had 3 eigenvalues above 1.0, that would be evidence that our
questions measured 3 underlying dimensions.

Modern techniques involve a "parallel analysis," in which the computer
generates random data (based on your data) but with *no* underlying
factors. Where your data differ from the simulated data, you have
evidence for a factor. We count the number of times that occurs, and
that is the number of factors you have.

___
Não vou explicar os detalhes da análise fatorial aqui (além do escopo deste curso), mas essencialmente analisa os padrões de associação entre os itens para determinar se eles "se agrupam" e, em caso afirmativo, como. Começamos com fa.parallel () do pacote psych. Dizemos que queremos usar o método do fator de minres (fm = 'minres'), que faz poucas suposições sobre a normalidade dos dados, e dizemos que queremos que a análise seja uma "análise fatorial" (fa = 'fa ').

Não vou entrar em detalhes sobre a análise, mas calculamos uma série de "autovalores" para nossos dados. Historicamente, o número de autovalores acima de 1,0 era visto como o número de fatores nos dados. Por exemplo, se tivéssemos 3 valores próprios acima de 1,0, isso seria uma evidência de que nossas perguntas mediram 3 dimensões subjacentes.

As técnicas modernas envolvem uma "análise paralela", na qual o computador gera dados aleatórios (com base em seus dados), mas sem fatores subjacentes. Onde seus dados diferem dos dados simulados, você tem evidências para um fator. Contamos o número de vezes que ocorre, e esse é o número de fatores que você possui.

```r
fa.parallel(dat, fm = 'minres', fa = 'fa')
```

Reliability
===========

Next, we can assess the reliability of the measure. Technically defined,
reliability is "the percentage of variance in the scores due to the
thing you're measuring" (i.e., due to the factor). If reliability is
100%, then our measure is perfectly reliable and every bump in the data
is real variation in sentiment. If reliability is less than 100%--for
example, 80%--then most of the variation is due to the thing you are
measuring but the remaining 20% is measurement error, other random
unrelated junk that you do not care about and is watering down your
measure.

I will point out that there are trade-offs in measurement. You can get a
perfectly reliable measure by asking the same question several different
ways, but that is redundant and frankly uninformative. It's worth it to
ask questions slightly differently (e.g., words such as "friendly" and
"awesome" are both positive words but not identical). You add more
information to your measure that way, but the trade off is that
reliability can suffer. In general, reliability between .70 and .90 is
good, with .80-.95 preferred.

To run our test, we use `alpha()` from the `psych` package. Since both
`ggplot2` and `psych` have commands with that name, it's useful to
preface the command with `psych::` to be sure that it pulls the command
from `psych`:

___
Confiabilidade
===
A seguir, podemos avaliar a confiabilidade da medida. Definida tecnicamente, confiabilidade é "a porcentagem de variância nas pontuações devido ao que você está medindo" (ou seja, devido ao fator). Se a confiabilidade for 100%, então nossa medida é perfeitamente confiável e cada aumento nos dados é uma variação real no sentimento. Se a confiabilidade for inferior a 100% - por exemplo, 80% - então a maior parte da variação é devido ao que você está medindo, mas os 20% restantes são erros de medição, outro lixo aleatório não relacionado com o qual você não se importa e é diluindo sua medida.

Vou salientar que existem compensações na medição. Você pode obter uma medida perfeitamente confiável fazendo a mesma pergunta de várias maneiras diferentes, mas isso é redundante e francamente não informativo. Vale a pena fazer perguntas de maneira um pouco diferente (por exemplo, palavras como "amigável" e "incrível" são palavras positivas, mas não idênticas). Você adiciona mais informações à sua medida dessa maneira, mas a desvantagem é que a confiabilidade pode ser prejudicada. Em geral, a confiabilidade entre 0,70 e 0,90 é boa, com preferência de 0,80 a 0,95.

Para executar nosso teste, usamos alpha () do pacote psych. Como ggplot2 e psych têm comandos com esse nome, é útil prefaciar o comando com psych :: para ter certeza de que puxa o comando de psych:
